Cruce de datos

Los 10 años con màs contaminaciòn ordenados de mayor a menor

```python
# Agrupar por año y calcular el promedio de emisiones de CO2
co2_by_year = df.groupby('Año')['CO2'].mean()

# Ordenar los resultados de mayor a menor
co2_by_year_sorted = co2_by_year.sort_values(ascending=False)

# Mostrar los años con más contaminación
print(co2_by_year_sorted.head(10))
```

Los 10 años con más ahorro de combustible

```python

# Agrupar por año y calcular el promedio de ahorro de combustible
ahorro_by_year = df.groupby('Año')['AhorroGasto'].mean()

# Ordenar los resultados de mayor a menor ahorro
ahorro_by_year_sorted = ahorro_by_year.sort_values(ascending=False)

# Mostrar los 10 años con más ahorro de combustible
print(ahorro_by_year_sorted.head(10))
```

Los 10 autos más contaminantes

```python
# Ordenar el DataFrame por emisiones de CO2 de mayor a menor
df_sorted = df.sort_values(by='CO2ColaGpm', ascending=False)

# Seleccionar los 10 modelos más contaminantes
top10_contaminantes = df_sorted[['Modelo', 'CO2ColaGpm']].head(10)

# Mostrar los resultados
print(top10_contaminantes)
```

Los 10 modelos menos contaminantes

```python
# Ordenar el DataFrame por emisiones de CO2 de menor a mayor
df_sorted = df.sort_values(by='CO2ColaGpm', ascending=True)

# Seleccionar los 10 autos menos contaminantes
bottom10_contaminantes = df_sorted[['Modelo', 'CO2ColaGpm']].head(10)

# Mostrar los resultados
print(bottom10_contaminantes)
```
