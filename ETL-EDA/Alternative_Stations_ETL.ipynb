Columna EV Pricing

Entre los valores que encontramos en la columna EV Pricing:

- Gratuito: Algunas estaciones de carga ofrecen el servicio de manera gratuita. 

- Por kWh (kilovatio-hora): Este es un esquema común donde se cobra una tarifa basada en la cantidad de electricidad consumida durante la carga. Los precios pueden variar, como "$0.13 por kWh" o "$0.35 por kWh".

- Por hora: Otra estructura de precios cobra por el tiempo que el vehículo está conectado a la estación de carga, sin importar cuánta electricidad consuma. Ejemplos incluyen "$1.50 por hora" o "$2 por hora".

- Por sesión: Algunas estaciones cobran una tarifa plana por cada sesión de carga, independientemente de la duración o la cantidad de energía consumida, como "$2 por sesión" o "$5 por sesión".

- Tarifas variables: Algunos esquemas de precios pueden incluir tarifas variables por estacionamiento y/o energía, indicando que el costo puede variar según ciertas condiciones, como la duración del estacionamiento o la cantidad de energía consumida.

- Donaciones: En algunos casos, la carga puede ser gratuita, pero se anima a los usuarios a hacer donaciones como una forma de apoyar el servicio.

- Tarifas complejas: Algunas estaciones utilizan estructuras de precios más complejas, combinando tarifas por kWh, por tiempo, y cargos por estacionamiento, lo que puede resultar en un cálculo más complicado del costo total de la carga.

La diversidad en las estructuras de precios refleja los diferentes modelos de negocio y estrategias de las entidades que operan las estaciones de carga, lo cual dificulta encontrar un factor común para agrupar los datos.

## III Tratamiento outliers y duplicados

### 1 Outliers


### EV Level2 EVSE Num (Número de Conectores de Nivel 2 EVSE)

- **Interpretación:** Los valores que son considerados outliers (por ejemplo, 16) indican estaciones de carga con un número relativamente alto de conectores de Nivel 2 EVSE disponibles. Mientras que la mayoría de las estaciones de carga pueden tener menos conectores, estos outliers representan estaciones de carga de gran capacidad, posiblemente ubicadas en áreas de alta demanda o diseñadas para servir a muchos vehículos simultáneamente.

### EV DC Fast Count (Conteo de Carga Rápida DC)

- **Interpretación:** Los valores outliers como 20 y 25 sugieren estaciones de carga que ofrecen una cantidad significativamente alta de puntos de carga rápida DC, mucho más que la media de las estaciones. Esto podría indicar centros de carga diseñados para un tráfico muy alto o para servir como hubs principales de carga en rutas clave o áreas metropolitanas donde la demanda de carga rápida es alta.

Los ourliers de ambas columnas resaltan estaciones de carga que son inusualmente bien equipadas para ofrecer servicios de carga, lo que no implica que haya un error en los datos.

```python
# Seleccionamos solo las columnas numéricas
df_numeric = df_ny.select_dtypes(include=['float64', 'int64'])

# Calcula el rango intercuartílico (IQR) para cada columna numérica
Q1 = df_numeric.quantile(0.25)
Q3 = df_numeric.quantile(0.75)
IQR = Q3 - Q1

# Identifica los outliers
outliers = ((df_numeric < (Q1 - 1.5 * IQR)) | (df_numeric > (Q3 + 1.5 * IQR)))

# Muestra los outliers
for column in outliers.columns:
    if outliers[column].any():
        print(f"Outliers in {column}:")
        print(df_numeric[column][outliers[column]])
        print()
```
