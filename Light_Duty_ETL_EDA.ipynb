Estadísticas descriptivas:

- Año_modelo: Los vehículos en el conjunto de datos varían desde 1991 hasta 2022. La media está en el año 2014, lo que indica que la mayoría de los vehículos son relativamente recientes.

- m/g_alt_ciudad (Millas por galón con combustible alternativo en ciudad): Hay una amplia gama de eficiencia en ciudad para los vehículos con combustible alternativo, con una media de alrededor de 38.4 m/g y una desviación estándar alta, lo que indica una gran variabilidad. De hecho, el máximo es extremadamente alto (800 m/g), lo que podría sugerir la presencia de outliers o datos erróneos.

- m/g_Alt_Carretera (Millas por galón con combustible alternativo en carretera): Similar a la eficiencia en ciudad, pero con una media ligeramente menor de 37.4 m/g. La variabilidad también es alta, como lo indica la desviación estándar.

- m/g_Alt_Combinada (Millas por galón con combustible alternativo combinado): Esta medida combina la eficiencia en ciudad y carretera, con una media de 72.1 m/g y una desviación estándar de 36.8, lo que muestra una dispersión significativa de los valores.

- m/g_Conv_Ciudad (Millas por galón con combustible convencional en ciudad): Los valores son menores que los del combustible alternativo, con una media de 22.8 m/g y una variabilidad también considerable.

- m/g_Conv_Carretera (Millas por galón con combustible convencional en carretera):La eficiencia en carretera es mejor que en ciudad para combustibles convencionales, con una media de 27.4 m/g.

- m/g_Conv_Combinada (Millas por galón con combustible convencional combinado): Un promedio de 27.8 m/g sugiere que la eficiencia combinada no es muy diferente de la eficiencia en carretera para combustibles convencionales.

- Autonomia_electrico (Autonomía solo eléctrica): Los vehículos eléctricos tienen un rango variado, desde tan solo 8 km hasta 520 km, con una media de 148 km. La desviación estándar es alta (128.7 km), lo que indica una gran variabilidad en la autonomía de los vehículos eléctricos en el conjunto de datos.

```python
# Filtrar las columnas numèricas que son ID para no incluirlas en las estadìsticas descriptivas
columnas_numericas = df.select_dtypes(include=['float64', 'int64']).columns
columnas_sin_id = [col for col in columnas_numericas if not col.lower().startswith('id')]

# Mostrar estadísticas descriptivas de las columnas numéricas 
estadisticas_descriptivas = df[columnas_sin_id].describe()
print(estadisticas_descriptivas)
```

BOXPLOTS

```python
# Lista de las columnas 
columnas = ['Año_modelo', 'm/g_alt_ciudad', 'm/g_Alt_Carretera', 'm/g_Alt_Combinada',
            'm/g_Conv_Ciudad', 'm/g_Conv_Carretera', 'm/g_Conv_Combinada', 'Autonomia_electrico']

# Configurar el tamaño de la figura 
plt.figure(figsize=(20, 10))

# Generar un gráfico de caja para cada columna
for i, columna in enumerate(columnas, 1):
    plt.subplot(2, 4, i)  # 2 filas, 4 columnas, posición i
    sns.boxplot(y=df[columna])
    plt.title(columna)
    plt.tight_layout()

plt.show()
```
